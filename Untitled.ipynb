```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Chargement des donnes
data = np.loadtxt("nuage_de_points.txt", skiprows=1)
x, y, z = data[:, 0], data[:, 1], data[:, 2]

# Rgression des moindres carrs (calcul des coefficients a, b, c)
X = np.column_stack((x, y, np.ones_like(x)))
coeff = np.linalg.lstsq(X, z, rcond=None)[0]

# Cration de la grille pour la surface ajuste
X_grid, Y_grid = np.meshgrid(np.linspace(min(x), max(x), 50), np.linspace(min(y), max(y), 50))
Z_grid = coeff[0] * X_grid + coeff[1] * Y_grid + coeff[2]

# Tracer les points et la surface ajuste
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, color='r', label="Points observs")
ax.plot_surface(X_grid, Y_grid, Z_grid, color='b', alpha=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Rgression des moindres carrs')
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Chargement des donnes
data = np.loadtxt("nuage_de_points.txt", skiprows=1)
x, y, z = data[:, 0], data[:, 1], data[:, 2]

# Cration de la matrice X (ajout d'une colonne de 1 pour le biais)
X = np.column_stack((x, y, np.ones_like(x)))  # Matrice des variables indpendantes
y = z  # Vecteur des valeurs observes

# Calcul des coefficients via la mthode des moindres carrs : (X^T * X)^-1 * X^T * y
XT = X.T  # Transpose de X
coeff = np.linalg.inv(XT @ X) @ XT @ y  # Rsolution du systme

# Les coefficients a, b, c sont dans 'coeff'
a, b, c = coeff[0], coeff[1], coeff[2]
print(f"Coefficients : a = {a}, b = {b}, c = {c}")

# Cration de la grille pour la surface ajuste
X_grid, Y_grid = np.meshgrid(np.linspace(min(x), max(x), 50), np.linspace(min(y), max(y), 50))
Z_grid = a * X_grid + b * Y_grid + c  # Surface ajuste

# Tracer les points et la surface ajuste
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, color='r', label="Points observs")
ax.plot_surface(X_grid, Y_grid, Z_grid, color='b', alpha=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Rgression des moindres carrs')
ax.legend()
plt.show()
```
