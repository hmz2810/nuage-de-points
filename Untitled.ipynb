```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Chargement des données
data = np.loadtxt("nuage_de_points.txt", skiprows=1)
x, y, z = data[:, 0], data[:, 1], data[:, 2]

# Régression des moindres carrés (calcul des coefficients a, b, c)
X = np.column_stack((x, y, np.ones_like(x)))
coeff = np.linalg.lstsq(X, z, rcond=None)[0]

# Création de la grille pour la surface ajustée
X_grid, Y_grid = np.meshgrid(np.linspace(min(x), max(x), 50), np.linspace(min(y), max(y), 50))
Z_grid = coeff[0] * X_grid + coeff[1] * Y_grid + coeff[2]

# Tracer les points et la surface ajustée
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, color='r', label="Points observés")
ax.plot_surface(X_grid, Y_grid, Z_grid, color='b', alpha=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Régression des moindres carrés')
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Chargement des données
data = np.loadtxt("nuage_de_points.txt", skiprows=1)
x, y, z = data[:, 0], data[:, 1], data[:, 2]

# Création de la matrice X (ajout d'une colonne de 1 pour le biais)
X = np.column_stack((x, y, np.ones_like(x)))  # Matrice des variables indépendantes
y = z  # Vecteur des valeurs observées

# Calcul des coefficients via la méthode des moindres carrés : (X^T * X)^-1 * X^T * y
XT = X.T  # Transposée de X
coeff = np.linalg.inv(XT @ X) @ XT @ y  # Résolution du système

# Les coefficients a, b, c sont dans 'coeff'
a, b, c = coeff[0], coeff[1], coeff[2]
print(f"Coefficients : a = {a}, b = {b}, c = {c}")

# Création de la grille pour la surface ajustée
X_grid, Y_grid = np.meshgrid(np.linspace(min(x), max(x), 50), np.linspace(min(y), max(y), 50))
Z_grid = a * X_grid + b * Y_grid + c  # Surface ajustée

# Tracer les points et la surface ajustée
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, color='r', label="Points observés")
ax.plot_surface(X_grid, Y_grid, Z_grid, color='b', alpha=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Régression des moindres carrés')
ax.legend()
plt.show()
```
